## Task 2

`[0.2]` Create a new column `whoWon` which contains `"red"` or `"blue"` based on the `blueWins` column.

````python
# Your code here
def decideWhoWon(row):
    if row['blueWins'] == 1:
        val = 'blue'
    else:
        val = 'red'
    return val

lol_data['whoWon'] = lol_data.apply(decideWhoWon, axis=1)
lol_data.sample(5)
````

`[0.2]` Create a new column `firstKill` which contains `"red"` or `"blue"` based on either `blueFirstBlood` or `redFirstBlood`.

````python
# Your code here
def whoKilledFirst(row):
    if row['blueFirstBlood'] == 1:
        val = 'blue'
    elif row['redFirstBlood'] == 1:
        val = 'red'
    else:
        val = 'deathless game'
    return val

lol_data['firstKill'] = lol_data.apply(whoKilledFirst, axis=1)
lol_data.head()
````

`[0.1]` Set the dtype of the `whoWon` and `firstKill` to categorical.

````python
# Your code here
mapSites = ['blue', 'red']
lol_data['whoWon'] = pd.Categorical(lol_data['whoWon'], categories=mapSites)
lol_data['firstKill'] = pd.Categorical(lol_data['firstKill'], categories=mapSites)
````

## Task 5

`[0.8]` Create a new variable `lol_data_group` where you groupby `whoWon` and `firstKill`, with the following aggregations:

- `blueTotalGold` - sum
- `redTotalGold` -  sum
- `blueKills` - mean
- `redKills` - mean
- `blueGoldPerMin` - max
- `redGoldPerMin` - min

````python
# Your code here
lol_data_group = lol_data.groupby(['whoWon', 'firstKill']).agg({'blueTotalGold': 'sum', 'redTotalGold': 'sum',
                                                               'blueKills': 'mean', 'redKills': 'mean',
                                                               'blueGoldPerMin': 'max', 'redGoldPerMin': 'min'})
lol_data_group
````

`[0.2]` **Rename** (use good function here) the columns of `lol_data_group` to `blueTotalGoldSum`, `redTotalGoldSum`, `blueKillsMean`, `redKillsMean`, `blueGoldPerMinMax`, `redGoldPerMinMin`.

````python
# Your code here
lol_data_group = lol_data_group.rename(columns={'blueTotalGold': 'blueTotalGoldSum',
                                                'redTotalGold': 'redTotalGoldSum',
                                                'blueKills': 'blueKillsMean', 'redKills': 'redKillsMean',
                                                'blueGoldPerMin': 'blueGoldPerMinMax',
                                                'redGoldPerMin': 'redGoldPerMinMin'})
lol_data_group
````

`[1.0]` Create a triangle (half square) seaborn `clustermap` of the `lol_data` correlations.

- Change the color palette to "mako"

*Hint: You can check stack overflow how to create masks using `numpy`.*

````python
# Your code here
corr = lol_data.corr()
mask = np.triu(np.ones_like(corr, dtype=np.bool))
f, ax = plt.subplots(figsize=(11, 9))

sns.heatmap(corr, mask=mask, cmap="mako", vmax=.3, center=0,
            square=True, linewidths=.25)
````
